## Рекурсия


### Реализация

```python
def factorial_recursive(n):
    if n == 1:
        return n
    else:
        return n*factorial_recursive(n-1)
```

## Бинарный поиск

### Реализация

```python
def simple_search(arr,value):
    j=1
    for i in range(len(arr)):
        print(j)
        j+=1
        if arr[i]==value:
            return i
def binary_search(arr,value):
    low_index=0
    high_index=(len(arr))-1
    i=1
    while low_index<=high_index:
        print(i)
        i+=1
        mid_index=(low_index+high_index)//2
        if arr[mid_index]==value:
            return mid_index
        if arr[mid_index]>value:
            high_index=mid_index-1
        else:
            low_index=mid_index+1
    return None
```

## Сортировки

### Пузурьком 

#### Реализация

```python
def bubble(array):
    for i in range(len(array)-1):
        for j in range(len(array)-i-1):
            if array[j] > array[j+1]:
                buff = array[j]
                array[j] = array[j+1]
                array[j+1] = buff
        print(array)
    
```

### Расческой

#### Реализация 

```python
def comb(array):
    gap = len(array)
    swaps = True
    while gap > 1 or swaps:
        gap = int(gap // 1.25)  
        swaps = False
        for i in range(len(array) - gap):
            j = i + gap
            if array[i] > array[j]:
                array[i], array[j] = array[j], array[i]
                swaps = True
   
```

### Быстрая

```python
def quicksort(array):
    if len(array) <= 1:
        return array
    else:
        q = random.choice(array)
        left = []
        midle = []
        right = []
        for elem in array:
            if elem < q:
                left.append(elem) 
            elif elem > q: 
                right.append(elem) 
            else: 
                midle.append(elem)
        return quicksort(left) + midle + quicksort(right)
arr=[124,-124,12425,0]
arr=quicksort(arr)
print(arr)
```

### Поиск в глубину

```python
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()
    visited.add(start)

    print(start)

    for next in graph[start] - visited:
        dfs(graph, next, visited)
    return visited



```

```python
def bfs(graph, start):
    visited, queue = set(), [start]
    while queue:
        vertex = queue.pop(0)
        if vertex not in visited:
            visited.add(vertex)
            queue.extend(graph[vertex] - visited)
    return visited
```
