```python
titanic_data_frame=pd.read_csv('titanic.csv')
agg_foo={'Pclass':['count']}
grdf=titanic_data_frame.groupby(['Survived']).agg(agg_foo)
grdf
```

```python
first_class_passengers=titanic_data_frame[titanic_data_frame['Pclass']==1]
first_class_passengers
```

```python
second_class_passengers=titanic_data_frame[titanic_data_frame['Pclass']==2]
second_class_passengers
```

```python
third_class_passengers=titanic_data_frame[titanic_data_frame['Pclass']==3]
third_class_passengers_survived=third_class_passengers[third_class_passengers['Survived']==True]
third_class_survived_precent=(len(third_class_passengers_survived)/len(third_class_passengers))*100
third_class_survived_precent
```

```python
(len(third_class_passengers_survived[third_class_passengers_survived['Sex']=='female'])/len(third_class_passengers_survived))*100
```
